# ***Function to calculate the pourcentage of matched based by ages***

```python
def pourcentage(x,y):
  totale=x+y
  pr=x*100/(x+y)
  return  pr
```

```python
t=projetcopy.groupby('age')['match'].value_counts()
print(t) 
```

```python
t[20]['No Match']
```

```python
def allper(h,col):
  vb=[]
  pourcentages=[]
  for i in col.unique():

    y=h[i]['Match']
    x=h[i]['No Match']
    p=pourcentage(x,y)
    vb.append(i)
    pourcentages.append(round((100-p),2))
  return vb,pourcentages  
```

```python
a,b=allper(t,projetcopy['age'])
print(a)
print(b)
```

```python
df=pd.DataFrame()
df['ages']=a
df['pourcentages']=b
df.sort_values(by=['pourcentages'],ascending=False)
```

```python
df.plot(x='ages', y='pourcentages', kind='bar',title='how important ages to have matchs')
plt.show()
```

# ***function to dictionnaire to dataframe***

```python
def dictonnaire(list,data_name):
  dict = {}
  for i in list:
    dict[i] = []
    dict[i].append(data_name.describe().loc["mean",i]) # dictionnaire
    data=pd.DataFrame(dict,index=['mean'])        #to datafram
  return data  
```
